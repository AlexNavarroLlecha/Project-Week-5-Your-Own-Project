### Create the columns for Back and Forwards

```python
def field_position(c):
    if c['Position'] in ['L', 'HO', 'PR', 'BR']:
        return 'Forward'
    elif c['Position'] in ['FH', 'SH', 'C', 'W', 'FB']:
        return 'Back'

df_ry['Field Position'] = df_ry.apply(field_position, axis=1)    

df_ry
```
